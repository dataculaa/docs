#### 2D Scatter plots
A scatter plot is a diagram where each value in the data set is represented by a dot. Matplotlib has a methid for 2D scatter plots which needs two arrays with the same dimention. One array is used for x-axis and another one for y-axis.

In machine learning, scatter plot is mostly used to show **_how much one variable is affected by another variable_**. Scatter plot is also used in geographical representation.

To find out how scatter plot works in matplotlib, let's take a look at one example. Here in below example, we create 2 vectors of size 1000, which $x,y \sim \mathcal{N}(\mu, \sigma)$ are gaussian or normal distribution.


```python
import numpy as np
import matplotlib.pyplot as plt

# numpy has random function set which generates random values
# np.random.randn(n,m) generates vector of size n,m with normal distribution
x = np.random.randn(1,1000)
y = np.random.randn(1,1000)

fig = plt.figure(figsize=(6,5))
ax = fig.add_axes([0,0,1,1,])

# scatter function or method on axes, plots scatter of x and y
ax.scatter(x,y)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('scatter plot of x & y')
plt.show()
```

`np.random.randn(n,m)` generates vector of size n,m with normal distribution.

As above figure shows, there is no corrolation between $x$ and $y$ which means these are independent variables.

To have more intuitive example, lets consider an example of **Automobile Data Set** from UCI Machine Learning Repository.

This dataset has many field which we select only two fields, horsepower and price. Logically, higher horsepower result in higher price of the car, so we expect from scatter plot that with increase of horsepower, price is increased as well. Total number of 30 samples are used in this example.

link of dataset: https://archive.ics.uci.edu/ml/datasets/automobile


```python
import numpy as np
import matplotlib.pyplot as plt

# datasets from Automobile Data Set dataset
horsepower = [111, 111, 154, 102, 115, 110, 110, 110, 140, 101, 101, 121, 121, 121, 
              182, 182, 182, 48, 70, 70, 68, 68, 102, 68, 68, 68, 102, 88, 145, 58]

price = [13495, 16500, 16500, 13950, 17450, 15250, 17710, 18920, 23875, 16430, 
        16925, 20970, 21105, 24565, 30760, 41315, 36880, 5151, 6295, 6575, 5572, 
        6377, 7957, 6229, 6692, 7609, 8558, 8921, 12964, 6479]

fig = plt.figure(figsize=(6,5))
ax = fig.add_axes([0,0,1,1,])
ax.scatter(horsepower,price)
ax.set_xlabel('horse power')
ax.set_ylabel('price of car')
ax.set_title('scatter plot of horse power vs. price')
ax.grid(color='r', linestyle='--', alpha=0.4)
```

As above figure shows, there is a relation (actually corrolation) between horsepower and price. So as stated before, scatter plots usually are used to show the corrolation of two variables in datasets.

BTW, because the price of car is not directly corrolated with horsepower (it depends to other parameters such as age, milage etc.), so here relation is not straight line. 

Note: all options such as grid(), set_title() & etc., for plots are also valid for scatter plots.

Now suppose that, we have received another batch of size 20 samples and we want to add to current scatter plot with different color.


```python
# add extra data to current scatter plot
horsepower2 = [92, 73, 92, 92, 92, 161, 161, 156, 156, 
              52, 85, 52, 85, 85, 68, 100, 90, 90, 110, 68]

price2 = [8948, 10698, 9988, 10898, 11248, 16558, 15998, 15690, 15750, 
          7775, 7975, 7995, 8195, 8495, 9495, 9995, 11595, 9980, 13295, 13845]

# adding new samples to scatter plot, automatically changes the color of new samples
ax.scatter(horsepower2,price2)
# ax.scatter(horsepower2,price2, color='r')
```

By adding new data to scatter plot, color of new samples will be automatically different from previous samples. Orange color in below figure, is related to new samples.

BTW, if you want to specify the color by yourself, it is possibe to use option `color = 'desired color'` as well.

In scatter plot, another option which might be useful, is size of each points specified by `s`. In the next example, we are going to check UCI Real estate valuation data set.

In this dataset, there are multiple columns of data, but we only use latitude, longitude and house price of unit area. We want to see how price per unit of a house is changing in different latitude & longitude. We only use 40 random sample of this dataset.

link of dataset: https://archive.ics.uci.edu/ml/datasets/Real+estate+valuation+data+set

```python
# Real estate valuation data set example
import numpy as np
import matplotlib.pyplot as plt

lat = [24.98298, 24.98034, 24.98746, 24.98746, 24.97937, 24.96305, 24.97933, 
       24.98042, 24.95095, 24.96731, 24.97349, 24.97433, 24.96515, 24.96108, 
       24.99156, 24.9824, 24.97744, 24.97544, 24.9675, 24.96772, 24.96314, 
       24.97528, 24.95204, 24.97528, 24.97353, 24.97542, 24.98085, 24.95593, 
       24.97419, 24.97563, 24.94826, 24.98281, 24.97349, 24.97841, 24.98419, 
       25.01459, 24.96386, 24.95204, 24.97201, 24.98203]

long = [121.54024, 121.53951, 121.54391, 121.54391, 121.54245, 121.51254, 
        121.53642, 121.54228, 121.48458, 121.51486, 121.53372, 121.5431, 
        121.53737, 121.51046, 121.53406, 121.54619, 121.54458, 121.53119, 
        121.54451, 121.54102, 121.51151, 121.54541, 121.54842, 121.54541, 
        121.53885, 121.51726, 121.54391, 121.53913, 121.53797, 121.54694, 
        121.49587, 121.53408, 121.53451, 121.54281, 121.54243, 
        121.51816, 121.51458, 121.54842, 121.54722, 121.54348]

price_per_unit = [37.9, 42.2, 47.3, 54.8, 43.1, 32.1, 40.3, 46.7, 18.8, 22.1, 
                  41.4, 58.1, 39.3, 23.8, 34.3, 50.5, 70.1, 37.4, 42.3, 47.7, 
                  29.3, 51.6, 24.6, 47.9, 38.8, 27, 56.2, 33.6, 47, 57.1, 22.1, 
                  25, 34.2, 49.3, 55.1, 27.3, 22.9, 25.3, 47.7, 46.2]


fig = plt.figure(figsize=(6,5))
ax = fig.add_axes([0,0,1,1,])

# opton s in scatter plot, specifies the size of each point
ax.scatter(lat,long,s = price_per_unit, alpha=1)
ax.set_xlabel('latitude')
ax.set_ylabel('longitude')
ax.set_title('price per unit')
```

So as above figure shows, the size of point varies with the house price per unit which shows the higher the longitude, the higher the price.

#### Line Plots and Basics
Plotting in matplotlib can be done with two methods

- Functional: simpler, less control over properties
- Object Oriented: advanced, more control over properties

### Functional approach

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

Let's plot exponential distribution function $f(x) = \lambda e^{-\lambda x} \text{ and } x \geq 0, \lambda \gt 0$. Let's suppose $\lambda = 0.5$ for this example. We use `numpy to generate` $x$.

```python
import numpy as np
x = np.linspace(0,10,100)
```

Which `np.linspace(0,10,100)` devide $range(0,10)$ to 100 points with equal distance. 
Then we create $f(x)$, let's call it $y$ for short notation.

```python
Lambda = 0.5
y = Lambda * np.exp(-Lambda * x)
```

Now let's plot $y$ versus $x$.

```python
plt.plot(x,y)
```

And here are more options for plot:

```python
# change shape and color of line plot 
plt.plot(x,y, '--r')

# add title, xlabel, ylabel
plt.xlabel('x label')
plt.ylabel('probability function')
plt.title('this is exponential function')
```

How we can plot multiple figures in the same canvas? It's possible by using `subplot` function from `pyplot` module.

`plt.subplot(#of rows, #of cols, targetId)` gets number of rows, number of cols and target subplot number and activate it for the next plot.


```python
# Create multiple plots on same canvas
plt.subplot(1,2,1)
plt.plot(x,y,'--b')

plt.subplot(1,2,2)
plt.plot(x,y,'--r')
```

A full list of plot() properties are available in below link: https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html

### Object Oriented Method approach

Using OO method just gives you more options and more feasibility. The main idea is to create an object `figure` and then just call methods (simply functions).


```python
# lets consider 2 different Gaussian distribution functions
x = np.linspace(-5,5,100)

SigmaOne = 1
MuOne = 2
const = 1/(SigmaOne * np.sqrt(2 * np.pi))
expo = -np.power(x-MuOne,2) / np.sqrt(2 * SigmaOne**2)
y1 =  const * np.exp(expo)

SigmaTwo = 5
MuTwo = 0
const = 1/(SigmaTwo * np.sqrt(2 * np.pi))
expo = -np.power(x-MuTwo, 2) /np.sqrt(2 * SigmaTwo**2)
y2 =  const * np.exp(expo)
```

Create an object of figure and then use methods(simply functions)

```python
# Consider fig as a blank canvas
fig = plt.figure()

# Add a set of axes to the empty figure. 
# The dimensions [left, bottom, width, height] of the new axes.
# All quantities are in fractions of figure width and height, range from 0 to 1.
ax = fig.add_axes([0.1, 0.1, 0.7, 0.7])

# Let's plot on the axes we just created
ax.plot(x,y1, '--g')

# Here is how to set labes and titles.
ax.set_xlabel('x values')
ax.set_ylabel('probability function')
ax.set_title('This is Gaussian function')
```

Fantastic! 
Now Let's create plot in plot using axes methods of figure.

First create two different axes and then add different data to each axes.

```python
# You can pass figsize(w,h) to change initial width and height in inches.
fig = plt.figure(figsize=(7,5))

# Add 2 set of axes to the empty figure. 
# order of box is [left, bottom, width, height]
ax1 = fig.add_axes([0.1, 0.1, 0.9, 0.9])
ax2 = fig.add_axes([0.2, 0.65, 0.3, 0.3])

# Plot different lines in each axes
ax1.plot(x,y1, '--r')
ax2.plot(x,y2, '--b')

ax1.set_title('Gaussian, SigmaOne = 1, MuOne = 2')
ax2.set_title('Gaussian, SigmaTwo = 5, MuTwo = 0')
```

Now let's create subplots using OO methods. Here ax and fig in below code, are called tuple unpacking. Actually subplots() directly add_axes() to the figure and then will name it as ax.

```python
# Create subplot in OO mehtods
# Here only one subplot exist
# here fig, ax is called tuple unpacking
# Calling subplots, actually directy add_axes to the figure as well
fig, ax = plt.subplots()

# use axes same as before, to plot
ax.plot(x,y1)
```

Using subplots, it is possible to create multiple subplots, So we need to specify the #Rows and #Cols. Here ax is an array of axes which can be updated by index.

To access different axes, just use related index as shown in above code (for example ax[1]).


```python
# create multiple subplots by specifying #Rows and #cols
fig, ax = plt.subplots(nrows=2, ncols=1)

# ax is now an array of axes, So we can iterate through it
# Let's now plot in each subplots
ax[0].plot(x,y1)
ax[0].set_title('top plot')

ax[1].plot(x,y2, '--r')

# move second title below to avoid text collisions
# y = -0.4 is a relative co-ordinates
ax[1].set_title('second plot', y=-0.4)
```

#### Figure size, ratio & DPI
To change size, aspect ratio and DPI (Dots Per Inches), for example in below code, initialize figure on creation.

```python
# Figure size, acpect ratio and DPI
# figsize accept width and height in inches
# dpi is dots per inches, 
fig = plt.figure(figsize=(6,5), dpi=100)

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y1)
```

Which gives you a bigger figure in size of (width, height) inches. You can conrol the quality of figure by using dpi option.
For subplots, the concept is same. 

```python
# dpi and figsize in subplots
fig, ax = plt.subplots(nrows=2, ncols=1, figsize=(8,6), dpi=100)

# ax is now an array of axes, So we can iterate through it
# Let's now plot in each subplots
ax[0].plot(x,y1)
ax[1].plot(x,y2)
```

#### Legends
Using legends, it is possible to describe each plot. To add legends, for each plot, `label` is used. Then finally legend() method is called to update the figure. Also location of legends is specified by `loc` option.

```python
# add legends to descibe plots using label in plot() method
fig = plt.figure(figsize=(6,4))
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y1, label = 'SigmaOne = 1, MuOne = 2')
ax.plot(x,y2, '--r', label = 'SigmaOne = 5, MuOne = 0')

# run legend() method to append legends to figure by specifyging location
# sample loc: best, upper right, upper left, center right, center, right & etc.
# or you can use tupple for location of legent loc=(0.1 , 0.1)
ax.legend(loc='upper left')
```

#### Colors, line options & grid
It is possible to change line apprearance to have a nice looking figure. multiple options are available for specifying color (line color), linewidth (tickness of line), alpha (transparency of line), linestyle (line pattern in plots).
Here is an example style with sample choices for each option in comments.

```python
# color: blue, red, purple, orange, black, ... Or Hex code such as #ff8f21
# linewidth: float number for example 2.1
# linestyle: '-', '--', '-.', ':'
# alpha: float range 0 to 1
fig = plt.figure(figsize=(6,4))
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y1, label = 'gaussian', color='#ff8f21', 
                                  linewidth=2.1, linestyle='-.', alpha=0.5)

# add grid to the axes for better readablity.
ax.grid(color='g', linestyle='-', linewidth=1, alpha=0.1)
```

#### Marker
Sometimes there is a need to highlight datapoints in the plots. marker option gives this possibility to specify data points with different style options.


```python
# mark data points in plot
x = np.linspace(0,10,20)
y = np.sqrt(x)

fig = plt.figure(figsize=(6,4))
ax = fig.add_axes([0,0,1,1])

# marker: o, +, *
# markersize: fload number > zero
# markerfacecolor & markeredgecolor: black, orange, red, blue, ##56ff98
# markeredgewidth: fload number > zero
ax.plot(x,y, label='sqrt(x)', color='purple',linewidth=0.8, marker='o', 
                                  markersize=10, markerfacecolor='orange', 
                                  markeredgecolor='black', markeredgewidth=2)

ax.grid(color='r', linestyle='--', linewidth=1, alpha=0.2)
```

#### Axis range
Axis ranges can be set or limited using methods on axes. Sometimes there is a need to focus on part of the plot to show ranges more clearly.
In below code, both horizontal and vertical axis ranges are changed.

```python
# contol axis ranges
x = np.linspace(-10,10,100)

Sigma = 3
Mu = 1
const = 1/(Sigma * np.sqrt(2 * np.pi))
expo = -np.power(x-Mu,2) / np.sqrt(2 * Sigma**2)
y =  const * np.exp(expo)

fig = plt.figure(figsize=(6,4))
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, '--')

# Set limitation on x and y axis
ax.set_xlim([-5,5])
ax.set_ylim([0,0.2])
ax.grid(color='r', linestyle='--', linewidth=1, alpha=0.2)
```

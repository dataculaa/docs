#### 2D Histogram plots
A histogram is mostly used to show frequency of distrubution which shows for example how many of observations are within the given intervals. Histogram are also used to find out the shape of distribution.

Let's create few samples of normal distribution and plot histogram to find out the shape of samples.


```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

Mu = 0
Sigma = 1
N = 1000 # Number of samples
x = np.random.normal(Mu, Sigma, N)

# Verification of Mu & Sigma: Calculate error to show as text in plot
Mu_err = abs(Mu - np.mean(x))
Sigma_err = abs(Sigma - np.std(x))
comment = 'Mean error is {:.2f} and Variance error is {:.2f}'.format(Mu_err, Sigma_err)

fig = plt.figure(figsize=(6,5))
ax = fig.add_axes([0,0,1,1])

# hist() method is used to plot histogram, bins is the number of chunks
# density=True adjust the magnitude to sum up to 1
count, bins, ignored = ax.hist(x, bins=20, density=True)

# Gaussian/Normal distribution
Nu = 1/(Sigma * np.sqrt(2 * np.pi)) * np.exp(-(bins-Mu)**2 / (2 * Sigma**2))
ax.plot(bins, Nu, '--r', linewidth=2)
ax.set_title('Histogam of samples from normal distribution')
ax.text(0.6, 0.4, comment)
ax.grid(color='g', alpha=0.1)
plt.show()
```

Note that np.random.normal(Mu, Sigma, N) generates $N$ samples with gaussian distribution of $\mu = Mu$ and $\sigma = Sigma$.

`hist()` method is used to plot histogram, with different options such as, bins which is the number of chunks, density=True which adjust the magnitude to sum up to 1 (To satisfy PDF properties).

Error for both mean and variance of generated samples are calculated to show the accuracy of samples.

hist(x, bins=20, density=True) is used to calculate histogram with 20 bins which seems quite well for this example. Then we plot normal distribution to compare with histogram. The result does match with negligible error.

Using `text()` method, it is possible to add text to plot convas in specified position ratios.
